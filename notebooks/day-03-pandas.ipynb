```python
df["Open"].tail() # it shows the last five rows of open column in the data frame
```

```python
df[["High", "Low"]].head(4) # it shows the open and close columns of first 4 rows in the data frame
```

```python
df["Close"] = pd.to_numeric(df["Close"], errors='coerce') # it converts the close column to numeric data type, if there are any non-numeric values, it will be replaced with NaN
df["Close"].mean() # it calculates the mean of the close column
```

```python
df["Close"].max() # it calculates the maximum value of the close column
```

```python
df["Close"].min().round(3) # it calculates the minimum value of the close column and rounds it to 3 decimal places
```

```python
df["Close"].describe().round(2) # it gives the summary statistics of the close column and rounds it to 2 decimal places
```

```python
df["Volume"] = pd.to_numeric(df["Volume"], errors='coerce').head() #it converts the volume column to numeric data type, if there are any non-numeric values, it will be replaced with NaN
df[df["Volume"] > 1000000] # it filters the rows where the volume is greater than 1000000
```

```python
df[df["Close"] > 500].tail() #it gives the info about the last 5 row where closing price is greater than 500
```

```python
df["Returns"] = df["Close"].pct_change() # it calculates the percentage change in the close column and creates a new column called returns
df.head()
#What is happening?
#.pct_change() = percentage change
#Automatically computes:
#(today - yesterday) / yesterday
```
